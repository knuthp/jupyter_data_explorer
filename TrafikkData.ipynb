````python
import requests
import pandas as pd
api_url = "https://www.vegvesen.no/trafikkdata/api/"
````

````python
def perform_query(query):
    headers = {"Content-Type" : "application/json"}
    res = requests.post(api_url, headers=headers, json={"query" : query})
    return res.json()
````

## Hourly

````python
def hourly_to_df(json):
    data = json['data']['trafficData']['volume']['byHour']['edges']
    df = pd.json_normalize(
        data=data,
        meta=[
            ['node', 'from'], ['node', 'to'], 
            ['node', 'total', 'volumeNumbers', 'volume'], 
            ["node", "total", "coverage", "percentage"],
        ],
        record_path=[
            ['node', 'byDirection'],        
    #        ['node', 'byLane'],
        ],
        sep="_",
    )
    df['node_from'] = pd.to_datetime(df['node_from'])
    df['node_to'] = pd.to_datetime(df['node_to'])
    return df
````

````python
import arrow

def get_local_datetime(date_str):
    return arrow.get(date_str, tzinfo="Europe/Oslo")

def get_one_day(date_str):
    date_from = get_local_datetime(date_str)
    date_to = date_from.shift(days=1)
    return (str(date_from), str(date_to))
````

````python
def one_day_hourly(reg_id, date_str):
    date = get_one_day(date_str)
    return f"""
    {{
      trafficData(trafficRegistrationPointId: "{reg_id}") {{
        volume {{
          byHour(from: "{date[0]}", to: "{date[1]}") {{
            edges {{
              node {{
                from
                to
                byDirection {{
                  heading
                  total {{
                    volumeNumbers {{
                      volume
                    }}
                    coverage {{
                      percentage
                    }}
                  }}
                }}
                byLane {{
                  lane {{
                    laneNumber
                    laneNumberAccordingToMetering                
                  }}
                  total {{
                    volumeNumbers {{
                      volume
                    }}
                    coverage {{
                      percentage
                    }}
                  }}            
                }}
                total {{
                  volumeNumbers {{
                    volume
                  }}
                  coverage {{
                    percentage
                  }}
                }}
              }}
            }}
          }}
        }}
      }}
    }}
    """

#one_day_hourly(point, "2020-09-01")
````

````python
def count_point_by_direction(point, start):
    query = one_day_hourly(point, start)
    json = perform_query(query)
    return hourly_to_df(json)
````

````python
import altair as alt
def chart_day(df):
    df['week']= df['node_from'].dt.isocalendar().week
    df['weekday_hour'] = df['node_from'].dt.day_name() + "_" + df['node_from'].dt.hour.astype(str).str.zfill(2)
    return alt.Chart(df).mark_bar().encode(
        x = 'node_from',
        y = 'total_volumeNumbers_volume',
        color = 'heading',
        tooltip = ['node_from', 'total_volumeNumbers_volume', 'heading']
    )
point = "00000V443295"
df = pd.concat([
    count_point_by_direction(point, "2020-09-12"),
    count_point_by_direction(point, "2020-09-13"),
    count_point_by_direction(point, "2020-09-14"),
    count_point_by_direction(point, "2020-09-15"),
    count_point_by_direction(point, "2020-09-16"),
    count_point_by_direction(point, "2020-09-17"),
    count_point_by_direction(point, "2020-09-18"),

    count_point_by_direction(point, "2020-09-19"),
    count_point_by_direction(point, "2020-09-20"),
    count_point_by_direction(point, "2020-09-21"),
    count_point_by_direction(point, "2020-09-22"),
    count_point_by_direction(point, "2020-09-23"),
    count_point_by_direction(point, "2020-09-24"),
    count_point_by_direction(point, "2020-09-25"),
])
chart_day(df)
````
